```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as matlab
```

```python
dataset=pd.read_csv("ad_click_dataset.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset.dropna(inplace=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['id', 'age', 'full_name_User100', 'full_name_User1000',
       'full_name_User10000', 'full_name_User1001', 'full_name_User1002',
       'full_name_User1011', 'full_name_User1012',
       'device_type_Tablet', 'ad_position_Side', 'ad_position_Top',
       'browsing_history_Entertainment', 'browsing_history_News',
       'browsing_history_Shopping', 'browsing_history_Social Media',
       'time_of_day_Evening', 'time_of_day_Morning', 'time_of_day_Night']]
dependent=dataset[['click']]
```

```python
independent.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy'],'max_features':['auto','sqrt','log2']}
grid=GridSearchCV(RandomForestClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,y_train)
```

```python
re=grid.cv_results_
grid_prediction=grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_prediction)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_prediction)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
